```python
if REMERGE:
    import glob
    import re
    import pandas as pd
    result = []
    DIR = "../simulation/competitive/thetas"
    for file in glob.glob(f"{DIR}/result*.csv"):
        match = re.search(r"result-(\d+)", file)
        if match:
            with open(f"{DIR}/log-{match.group(1)}.txt") as log:
                m = re.search(r"thetaS=([\d\.]+)")
                if m:
                    thetaS = float(m.group(1))
                    data = pd.read_csv(file)
                    data["thetaS"] = thetaS
                    result.append(data[["thetaS", "BP3.PP", "BP3.PS", "BP3.RLP", "BP3.RLS", "BP3.PTXP", "BP3.PTXS"]])
    if result:
        pd.concat(result).to_csv(f"{DIR}/main.csv", index=None)
```

```python
data = pd.read_csv("../simulation/competitive/thetas/main.csv")
```
