```python
from plotter import *
```

```python
path = "../simulation/generate/0505-summary/main.csv"
data = pd.read_csv(path)
BINS = 100
CompetitiveKey = "BP2"
SelfishKey = "BS2"
```

```python
GAP = 0.01
```

```python
gapptx = data["BS2.PTXP"] - data["NCP.PTX"]
sameptx = data[(gapptx < GAP) & (gapptx > -GAP)]
```

```python
ecdfs(sameptx, ["BS2.RLP", "NCP.RL"], labels=["A", "B"])
```

```python
draw(
    sameptx["NCP.RL"] - sameptx[f"{SelfishKey}.RLP"],
    bins_width=0.05,
    grid=False,
    cumulative=False,
    xlabel=r"$R_L^{(P-NC)} - R_L^{(P-SF)}$",
    ylabel=None,
    xlim=[-0.8, 0.8]
)
```

```python
gapptx2 = data["BP2.PTXP"] - data["NCP.PTX"]
sameptx2 = data[(gapptx2 < GAP) & (gapptx2 > -GAP)]
```

```python
draw(
    sameptx2["NCP.RL"] - sameptx2[f"{CompetitiveKey}.RLP"],
    bins_width=0.05,
    grid=False,
    cumulative=True,
    xlabel=r"$R_L^{(P-NC)} - R_L^{(P-CP)}$",
    ylabel=None,
    xlim=[-0.8, 0.8]
)
```
