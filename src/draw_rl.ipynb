```python
from numpy import random
from params import *
from plotter import plot_3d
```

```python
def run(params: ProblemParams, custom={}, contour=False, surface=True, invertX=False, invertY=False):
    if not custom:
        custom =  {
            'Dpp': random.randint(1, 31),
            'Dss': random.randint(1, 31),
            'Dps': random.randint(1, 31),
            'Dsp': random.randint(1, 31),
            'Dpe': random.randint(1, 31),
            'Dpf': random.randint(1, 31),
            'Dse': random.randint(1, 31),
            'Dsf': random.randint(1, 31),
        }
    params.set_params(dynamic=True, **custom)
    print({k: v for k, v in params.get_state().items() if k.startswith("D")})
    args = {
        "minX": 0.0,
        "maxX": 20.0,
        "minY": 0.0,
        "maxY": 20.0,
        "minZ": 0.0,
        "maxZ": 1.0,
        "contour": contour,
        "surface": surface,
        "invertX": invertX,
        "invertY": invertY,
        "invertZ": False,
        "xlabel": "$p_P$", 
        "ylabel": "$p_S$", 
        "zlabel": "$R_L^{(P)}$"
    }
    plot_3d(params.leakage_rate_co, **args)
    plot_3d(lambda pp, ps: params.leakage_rate_co(pp, ps, side=Side.S), **args)
```

```python
p1 = {'Dpp': 15, 'Dss': 15, 'Dps': 9, 'Dsp': 28, 'Dpe': 8, 'Dpf': 6, 'Dse': 28, 'Dsf': 9}
p2= {'Dpp': 22, 'Dss': 8, 'Dps': 14, 'Dsp': 28, 'Dpe': 22, 'Dpf': 16, 'Dse': 15, 'Dsf': 7}

params = ProblemParams(PP=20, PS=20, NP=1, NS=1, rhoP=0.9, rhoS=0.5, RbP=1.0, RbS=1.0, R0=30)
run(params, custom=p1, contour=False, surface=True, invertX=False, invertY=True)
```
